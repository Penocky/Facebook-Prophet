```python
import pandas as pd
from fbprophet import Prophet
```

```python
#import saturating_forecasts data
df = pd.read_csv('data/example_wp_log_R.csv')
df.head()                 
```

```python
# specify the carrying capacity in a column cap
df['cap'] = 8.5
```

```python
# create new Prophet object and pass in the historical dataframe
m = Prophet(growth='logistic')
m.fit(df)
```

```python
# Extend dataframe into the future 3 years using the helper method Prophet.make_future_dataframe.
future = m.make_future_dataframe(periods = 1826)
future['cap'] = 8.5
future.tail()
```

```python
# Get predicted value yhat for future values with uncertainty intervals
forecast = m.predict(future)
#plot forecast
fig1 = m.plot(forecast)
```

```python
#Use the Prophet.plot_components method to see the forecast components: 
#the trend, yearly seasonality, and weekly seasonality of the time series.
fig2 = m.plot_components(forecast)
```

```python
#handle a saturating minimum with a column floor 
#similar way to the cap column specifies the maximum
df['y'] = 10 - df['y']
df['cap'] = 6
df['floor'] = 1.5
future['cap'] = 6
future['floor'] = 1.5
m = Prophet(growth='logistic')
m.fit(df)
fcst = m.predict(future)
fig = m.plot(fcst)
```
